Let's now group our qualities into 3 categories: low, medium and high quality wines.

<span style="font-size: larger; color: white;">White</span>

```python
white_copy['quality_label'] = 'high'
white_copy.loc[white_copy.quality < 5, 'quality_label'] = 'low'
white_copy.loc[(white_copy.quality >= 5) & (white_copy.quality <=7), 'quality_label'] = 'medium'
#df.loc[df['existing_column'] > 15, 'new_column'] = 'high_value'

display(white_copy.head())
```

<span style="font-size: larger; color: red;">Red</span>

```python
red_copy['quality_label'] = red_copy.quality.apply(lambda value: 'low' if value <= 5 else 'medium' if value <= 7 else 'high')
display(red_copy.head())
```

Create a new Dataframe combining both White and Red. White_red_df

```python
white_red_df = pd.concat([white_copy, red_copy])
display(white_red_df.shape)
print(white_red_df.head())
```
